### Importando pacote de vizualização

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set() 
```

### Bar Chart for Categorical Features
- Pclass
- Sex
- SibSp ( # de irmãos e cônjuge)
- Parch ( # de pais e filhos)
- Embarked
- Cabin

```python
def bar_chart(feature):
    survived = train[train['Survived']==1][feature].value_counts()
    dead = train[train['Survived']==0][feature].value_counts()
    df = pd.DataFrame([survived,dead])
    df.index = ['Survived','Dead']
    df.plot(kind='bar',stacked=True, figsize=(10,5))
```

```python
bar_chart('Sex')
```

O gráfico confirma que **Women** tem mais probabilidade de sobreviver do que **Men**

```python
bar_chart('Pclass')
```

O gráfico confirma **1st class** mais provavelmente sobreviveu do que **other classes**
O gráfico confirma **3rd class** mais provavelmente morto que **other classes**

```python
bar_chart('SibSp')
```

O gráfico confirma **uma pessoa a bordo com mais de 2 irmãos ou cônjuge** mais provavelmente sobreviveu
O gráfico confirma **uma pessoa a bordo sem irmãos ou cônjuge** mais provavelmente morto

```python
bar_chart('Parch')
```

O gráfico confirma **uma pessoa a bordo com mais de 2 pais ou filhos** mais provavelmente sobreviveu
O gráfico confirma **uma pessoa a bordo sozinha** mais provavelmente morta

```python
bar_chart('Embarked')
```

- O gráfico confirma **uma pessoa a bordo de C** sobreviveu com um pouco mais de probabilidade
- O gráfico confirma **uma pessoa a bordo de Q** mais provavelmente morta
- O gráfico confirma **uma pessoa a bordo de S** mais provavelmente morta
